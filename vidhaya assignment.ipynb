```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
URL="https://www.flipkart.com/search?q=t+shirts&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_7_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_7_0_na_na_na&as-pos=7&as-type=TRENDING&suggestionId=t+shirts&requestId=7ce718ec-d936-4d4e-bc64-e82375abdbcb"
```

```python
scarp=requests.get(URL)
```

```python
scarp
```

```python
link=BeautifulSoup(scarp.text)
```

```python
link
```

```python
products=link.select("._2WkVRV")
```

```python
products
```

```python
rate=link.select("._3bPFwb ._30jeq3")
```

```python
rate
```

```python
import pandas
```

```python
products=pandas.DataFrame(products,columns=["name"])
```

```python
products
```

```python
products["Price"]=rate
```

```python
products
```

```python
products.to_csv("D:/vidhya assign/assignment.csv")
```
